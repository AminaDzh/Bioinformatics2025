### 1.1 Чтение таблицы 

```python
directory = "/home/adzhioeva/homework_bioinf/hw_2/TRB_10_7"
filename = "TRB_10_7.tsv"  
path = os.path.join(directory, filename)
df = pd.read_csv(path, sep='\t')
df
```

### 1.2 Построение гистрограммы

```python
import matplotlib.pyplot as plt
```

```python
df['length'] = df['TGTGCCAGCAGCGGTAGGGGGACACTATTGGGCACCTTC'].apply(len)
plt.figure(figsize=(10, 6))  # Размер графика
plt.hist(df['length'], bins=30, color='blue', edgecolor='black')  # bins — количество столбцов
plt.title('Гистограмма длины прочтений')  # Заголовок
plt.xlabel('Длина прочтения')  # Подпись оси X
plt.ylabel('Частота')  # Подпись оси Y
plt.grid(True)  # Сетка
plt.show()
```
